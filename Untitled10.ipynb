```python
import spacy


nlp = spacy.load("en_core_web_sm")


with open('/content/sample_data/wikisent2.txt', 'r') as file:
    text_data = file.readlines()


for index, sentence in enumerate(text_data):
    sentence = sentence.strip()
    if not sentence:
        continue

    print(f"Sentence {index + 1}: {sentence}")


    doc = nlp(sentence)


    for token in doc:
        print(f"Word: {token.text}, POS: {token.pos_}")

    print("\n" + "="*50 + "\n")
```
